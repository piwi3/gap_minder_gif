```python
co2.columns = co2.columns.astype(int)
co2.columns
```

```python
co2.fillna(0, inplace = True)

def replace_mu(x):
    if str(x)[-1] == 'Âµ':
        return float(str(x[0:-1])) / (10 ** 6)
    else:
        return float(x)

for year in co2.columns:
    co2[year] = co2[year].apply(lambda x : replace_mu(x))
co2.info()
```

```python
co2.head() #check whether all entries are o.k. now
```

```python
# Loading income data
income = pd.read_csv('../data/gapminder_income_per_person_gdppercapita_ppp_inflation_adjusted.csv', index_col = 0)
income.head()
```

```python
income.columns = income.columns.astype(int)
income.columns
```

```python
income.shape
```

```python
#eliminate "k"s

def replace_k(x):
    x_new = x
    if str(x)[-1] == 'k': 
        x_new = 1000 * float(str(x).replace("k" , ""))
    return int(x_new)

for year in income.columns:
    income[year] = income[year].apply(lambda x : replace_k(x))

income.head()
```

## Calculate cumulated CO2 

```python
df_stacked = df.pivot(index = 'country', columns = 'year')
co2_total = df_stacked['co2'] * df_stacked['population']
co2_cum = co2_total.cumsum(axis=1)
```

```python
co2_cum
```

```python
co2_cum = co2_cum.reset_index().melt(id_vars = 'country', var_name = 'year', value_name = 'co2_cumulated')
co2_cum.head() 
```

```python
df = df.merge(co2_cum)
```

```python
df
```

# Produce figures and gif

```python
year = 2015
df_subset = df.loc[df['year'] == year]
df_subset.head()
```

```python
#df['continent'] = pd.Categorical(df['continent'])

from seaborn import xkcd_rgb, xkcd_palette, palplot
neon_palette = xkcd_palette([col for col in xkcd_rgb.keys() if col.startswith('neon')])

plt.style.use("dark_background")

plt.figure(figsize=(8,8))
ax = sns.scatterplot(x = 'income', 
                y = 'co2_cumulated',
                size = 'population',
                sizes = (20, 1000),
                hue = 'continent',
                data = df_subset, 
                alpha = 0.6, legend = True,
                palette = neon_palette,
                )
ax.set_title(f'Income vs. cumulated CO2 - {year}\n', fontsize=15)
ax.set_xlabel('Income per person')
ax.set_ylabel('CO2 emissions in tons cumulated')
ax.set_xlim([500,120_000])
ax.set_ylim([0,3.5*10**11])
ax.set(xscale="log")
ax.set_xticks([1000, 10000, 100_000])
ax.set_xticklabels(['1000', '10000', '100000'])
handles, labels = ax.get_legend_handles_labels()
for i, j in enumerate(labels):
    if j == 'continent':
          labels[i] = 'Continent'
    if j == 'population':
        labels[i] = '\nSize of population'
ax.legend(handles, labels, loc='upper left', frameon = False, ncol = 1)
# plt.text(x=40000, y= (1.5 * 10 ** 11), s= f'{year}',fontsize=28, alpha = 0.4)
sns.despine()
```

```python
for year in range(1960, 2016):
    df_subset = df.loc[df['year'] == year]
    
    plt.figure(figsize=(8,8))
    ax = sns.scatterplot(x = 'income', 
                y = 'co2_cumulated',
                size = 'population',
                sizes = (20, 1000),
                hue = 'continent',
                data = df_subset, 
                alpha = 0.6, legend = True,
                palette = neon_palette,
                )
    ax.set_title(f'Income vs. cumulated CO2 - {year}\n', fontsize=15)
    ax.set_xlabel('Income per person')
    ax.set_ylabel('CO2 emissions in tons cumulated')
    ax.set_xlim([500,120_000])
    ax.set_ylim([0,3.5*10**11])
    ax.set(xscale="log")
    ax.set_xticks([1000, 10000, 100_000])
    ax.set_xticklabels(['1000', '10000', '100000'])
    ax.legend(handles, labels, loc='upper left', frameon = False) # Using handles, labels from above
    #plt.text(x=40000, y= (1.5 * 10 ** 11), s= f'{year}',fontsize=28, alpha = 0.4)
    sns.despine()
    
    
    f_name = 'scatter_income_co2cum_' + str(year) + '.png'
    plt.savefig('./scatter_pics3/' + f_name, dpi = 250)
    plt.close()
    
```

```python
images = []
for i in range(1960,2016):
    filename = f'./scatter_pics3/scatter_income_co2cum_{i}.png'
    images.append(imageio.imread(filename))
    
imageio.mimsave('../images/gapminder_income_co2cum.gif', images, fps = 20)
```
